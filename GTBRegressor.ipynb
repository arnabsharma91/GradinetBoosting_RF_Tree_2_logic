```python
from sklearn.ensemble import GradientBoostingRegressor
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
import pandas as pd
import numpy as np
from joblib import dump, load
from sklearn import model_selection
from sklearn.metrics import accuracy_score
import csv as cv
```

```python
df = pd.read_csv('TestOracleRegression.csv')
data = df.values
X = data[:, :-1]
y = data[:,-1]
model = GradientBoostingRegressor(n_estimators=3)
model =model.fit(X, y)
```

```python
model.estimators_.shape
```

```python
fig = plt.figure(figsize=(15, 5))
plot_tree(model.estimators_[0, 0], 
          filled=True, impurity=False, 
          rounded=True)
```

```python
fig = plt.figure(figsize=(15, 5))
plot_tree(model.estimators_[1, 0], 
          filled=True, impurity=False, 
          rounded=True)
```

```python
fig = plt.figure(figsize=(15, 5))
plot_tree(model.estimators_[2, 0], 
          filled=True, impurity=False, 
          rounded=True)
```
